```python
#Import the requests library and define your OpenWeatherMap API key
import requests
import pandas as pd
import matplotlib.pyplot as plt

#Define your API key
api_key = 'bdc812c40cc9c634fcd7454e6631f86d'

#List of cities
cities = ['Bogota,CO', 'Medellin,CO', 'Cali,CO', 'Barranquilla,CO', 'Cartagena,CO', 'Bucaramanga,CO', 'Pereira,CO', 'Ibague,CO', 'Manizales,CO', 'Popayan,CO', 'Pasto,CO']

#Initialize lists to store data
city_names = []
temperatures = []
humidities = []

#Iteration over the list of cities
for city in cities:
    # Make a request to the API to get the weather data for the current city
    url = f'http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric'
    response = requests.get(url)
    data = response.json()

    #Save temperature and humidity data to lists
    city_names.append(data["name"])
    temperatures.append(data["main"]["temp"])
    humidities.append(data["main"]["humidity"])

#Create a DataFrame with the data
df = pd.DataFrame({
    'City': city_names,
    'Temperature (°C)': temperatures,
    'Humidity (%)': humidities
})

#Show data as table
print(df)
print()

#Show temperature graph
plt.figure(figsize=(10, 6))
plt.bar(df['City'], df['Temperature (°C)'], color='skyblue')
plt.title('Temperatures in Cities')
plt.xlabel('City')
plt.ylabel('Temperature (°C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
print()

#Show humidity graph
plt.figure(figsize=(10, 6))
plt.bar(df['City'], df['Humidity (%)'], color='lightgreen')
plt.title('Humidity in Cities')
plt.xlabel('City')
plt.ylabel('Humidity (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
