Calcular el valor de la pendiente de los valores dados en (x) y (y) con las formulas de pendiente e interseccion

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data
x_values = np.array([1, 2, 2, 3, 4, 4, 5, 6])
y_values = np.array([2, 3, 4, 4, 4, 6, 5, 7])

# Convert x and y to DataFrames
x = pd.DataFrame(x_values, columns=['x'])
y = pd.DataFrame(y_values, columns=['y'])

# Calculate xy and x squared
xy = x_values * y_values
x_squared = x_values**2

# Calculate the sum of each column
sum_x = np.sum(x_values)
sum_y = np.sum(y_values)
sum_xy = np.sum(xy)
sum_x_squared = np.sum(x_squared)

# Calculate the slope (m) and intercept (b) of the regression line
n = len(x_values)
m = (n*np.sum(xy) - np.sum(x_values)*np.sum(y_values)) / (n*np.sum(x_squared) - np.sum(x_values)**2)
b = (np.sum(y_values) - m*np.sum(x_values)) / n

# Combine individual values and sums
data = {'x': np.concatenate((x_values, [sum_x])),
        'y': np.concatenate((y_values, [sum_y])),
        'xy': np.concatenate((xy, [sum_xy])),
        'x^2': np.concatenate((x_squared, [sum_x_squared]))}

# Create a DataFrame for the table
df = pd.DataFrame(data, index=np.concatenate((np.arange(len(x_values)), ['Sum'])))

# Display the table
print("Table of Values:")
print(df)

# Create the regression line
regression_line = m * x_values + b

# Plot the points and the regression line
plt.scatter(x_values, y_values, color='blue', label='Data')
plt.plot(x_values, regression_line, color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()

# Print the slope and intercept
print(f"Slope (m): {m}, Intercept (b): {b}")
```

Calcular el valor de la pendiente usando la funcion np.polyfit

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Data
x_values = np.array([1, 2, 2, 3, 4, 4, 5, 6])
y_values = np.array([2, 3, 4, 4, 4, 6, 5, 7])

# Convert x and y to DataFrames
x = pd.DataFrame(x_values, columns=['x'])
y = pd.DataFrame(y_values, columns=['y'])

# Calculate xy and x squared
xy = x_values * y_values
x_squared = x_values**2

# Calculate the sum of each column
sum_x = np.sum(x_values)
sum_y = np.sum(y_values)
sum_xy = np.sum(xy)
sum_x_squared = np.sum(x_squared)

# Fit a linear regression using np.polyfit
coefficients = np.polyfit(x_values, y_values, 1)
slope = coefficients[0]
intercept = coefficients[1]

# Combine individual values and sums
data = {'x': np.concatenate((x_values, [sum_x])),
        'y': np.concatenate((y_values, [sum_y])),
        'xy': np.concatenate((xy, [sum_xy])),
        'x^2': np.concatenate((x_squared, [sum_x_squared]))}

# Create a DataFrame for the table
df = pd.DataFrame(data, index=np.concatenate((np.arange(len(x_values)), ['Sum'])))

# Display the table
print("Table of Values:")
print(df)

# Create the regression line
regression_line = slope * x_values + intercept

# Plot the points and the regression line
plt.scatter(x_values, y_values, color='blue', label='Data')
plt.plot(x_values, regression_line, color='red', label='Regression Line')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.legend()
plt.show()

# Print the slope and intercept
print(f"Slope (m): {slope}, Intercept (b): {intercept}")
```
